```python
X = new_dataset.drop(columns='is_fraud', axis=1)
Y = new_dataset['is_fraud']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
import pandas as pd
X_train = pd.get_dummies(X_train)
```

```python
# Ensure the 'trans_date_trans_time' column exists in X_train and X_test before processing
if 'trans_date_trans_time' in X_train.columns:
	X_train['trans_date_trans_time'] = pd.to_datetime(X_train['trans_date_trans_time'])
	X_train['year'] = X_train['trans_date_trans_time'].dt.year
	X_train['month'] = X_train['trans_date_trans_time'].dt.month
	X_train['day'] = X_train['trans_date_trans_time'].dt.day
	X_train = X_train.drop(columns=['trans_date_trans_time'])  # Drop original date column

if 'trans_date_trans_time' in X_test.columns:
	X_test['trans_date_trans_time'] = pd.to_datetime(X_test['trans_date_trans_time'])
	X_test['year'] = X_test['trans_date_trans_time'].dt.year
	X_test['month'] = X_test['trans_date_trans_time'].dt.month
	X_test['day'] = X_test['trans_date_trans_time'].dt.day
	X_test = X_test.drop(columns=['trans_date_trans_time'])  # Drop original date column
```

```python
print(X_train.isnull().sum())
```

```python
X_train = X_train.fillna(0) 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y_train = le.fit_transform(Y_train)
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# Ensure the test set has the same columns as the training set
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
import numpy as np
import joblib
import pandas as pd

model = joblib.load('credit_card_fraud_detection.joblib')

feature_names = [
    'trans_date_trans_time', 'cc_num', 'merchant', 'category', 'amt', 'first', 'last', 
    'gender', 'street', 'city', 'state', 'zip', 'lat', 'long', 'city_pop', 'job', 'dob', 
    'trans_num', 'unix_time', 'merch_lat', 'merch_long'
]

input_values = []

print("Enter values for the following features:")
for feature in feature_names:
    value = input(f"{feature}: ")
    input_values.append(value)

numeric_features = ['amt', 'zip', 'lat', 'long', 'city_pop', 'unix_time', 'merch_lat', 'merch_long']
for i, feature in enumerate(feature_names):
    if feature in numeric_features:
        input_values[i] = float(input_values[i])

input_df = pd.DataFrame([input_values], columns=feature_names)
input_df[numeric_features] = input_df[numeric_features].apply(pd.to_numeric)

input_df = input_df.reindex(columns=X_train.columns, fill_value=0)

prediction = model.predict(input_df)

if prediction[0] == 1:
    print("The transaction is predicted to be FRAUDULENT.")
else:
    print("The transaction is predicted to be NON-FRAUDULENT.")
```
